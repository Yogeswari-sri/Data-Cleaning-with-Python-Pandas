# **Data Cleaning and Transformation**

```python
df.drop('Useless_Col', axis=1,inplace=True)
```

```python
df.duplicated().sum()
```

```python
#Display Duplicate Rows
df[df.duplicated(keep=False)]
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

step 1 : First column Employee_ID

```python
df['Employee_ID']=df["Employee_ID"].astype(str)
```

Name column

```python
df["Name"] =df["Name"].str.title().str.strip()
```

Age column

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
# df["Age"]=df["Age"].fillna(df["Age"].mean())
```

Department column

```python
df["Department"].unique()
```

```python
df["Department"].value_counts()
```

```python
# Upper conversion
df["Department"] = df["Department"].str.upper()
```

```python
#Replace multiple values in a column
df['Department'].replace({
    'FINANACE': 'Finance',
    'MARKETING': 'Marketing',
    'SALES': 'Sales'
}, inplace=True)
```

Position column

```python
df["Position"].unique()
```

```python
df["Position"].value_counts()
```

```python
# Symbol conversion using Regular Expression
df["Position"] =df["Position"].str.replace(r'[^a-zA-Z]'," ",regex= True).str.strip()
```

Salary column

```python
df["Salary"].unique()
```

```python
df["Salary"].value_counts()
```

```python
# If duplicate values is there then only used this dropna
#df["Salary"].value_counts(dropna = False)
```

```python
#Replace "70k" to "70000"
df["Salary"] = df['Salary'].str. replace("70k","70000")
```

```python
# Remove only commas using Regular expression
df["Salary"] = df["Salary"].replace(r',', '', regex=True)

# Convert to float
df["Salary"] = df["Salary"].astype(float)
```

```python
df
```

```python
# Hierarchical imputation
# Level 1: Department + Position mean
dept_pos_mean = df.groupby(['Department','Position'])['Salary'].transform('median')

# # # # Level 2: Department mean
dept_mean = df.groupby('Department')['Salary'].transform('mean')

# # # Level 3: Overall mean
overall_mean = df['Salary'].mean()

# Hierarchical imputation
df['Salary'] = (df['Salary'].fillna(dept_pos_mean).fillna(dept_mean).fillna(overall_mean))
```

Feedback column

```python
df["Feedback"].unique()
```

```python
df["Feedback"].value_counts()
```

```python
# Remove symbol using Regular expression
df["Feedback"] = df["Feedback"].replace(r'"?",''', regex=True).str.title()
```

```python
df["Feedback"] = df['Feedback'].fillna("Average")
```

Has_car

```python
df["Has_Car"].unique()
```

```python
df["Has_Car"].value_counts()
```

```python
df['Has_Car'].replace({
    '0': 'No',
    '1': 'Yes'
}, inplace=True)
```

```python
df["Has_Car"] = df['Has_Car'].fillna("No")
```
